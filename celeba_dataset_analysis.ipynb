```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Put the path to your copy of celebA
full_df = pd.read_csv('celebA/list_attr_celeba.csv')
```

```python
male_df = full_df.loc[full_df['Male']==1]
female_df = full_df.loc[full_df['Male'] == -1]
pale_df = full_df.loc[full_df['Pale_Skin'] == 1]
unpale_df = full_df.loc[full_df['Pale_Skin'] == -1]

attractive_df = full_df.loc[full_df['Attractive'] == 1]
unattractive_df = full_df.loc[full_df['Attractive'] == -1]
```

```python
male_attract = male_df.loc[male_df['Attractive'] == 1]
female_attract = female_df.loc[female_df['Attractive'] == 1]

print(f'% of attractive males: {len(male_attract)/len(male_df)}')
print(f'% of attractive females: {len(female_attract)/len(female_df)}')
```

```python
pale_attract = pale_df.loc[pale_df['Attractive'] == 1]
unpale_attract = unpale_df.loc[unpale_df['Attractive'] == 1]

print(f'% of attractive pales: {len(pale_attract)/len(pale_df)}')
print(f'% of attractive unpales: {len(unpale_attract)/len(unpale_df)}')
```

```python
# Supposedly there is not a drastic picture quality difference between pale and unpale people
blurry_pale = pale_df.loc[pale_df['Blurry'] == 1]
blurry_unpale = unpale_df.loc[unpale_df['Blurry'] == 1]

print(f'% of blurry pales: {len(blurry_pale)/len(pale_df)}')
print(f'% of blurry unpales: {len(blurry_unpale)/len(unpale_df)}')
```

```python
old = full_df.loc[full_df['Young'] == -1]
old_attractive = old[old['Attractive'] == 1]

print(f'% of old attractive: {len(old_attractive)/len(old)}')
old_female = female_df[female_df['Young'] == -1]
old_male = male_df[male_df['Young'] == -1]

print(f'% males old: {len(old_male)/len(male_df)}')
print(f'% females old: {len(old_female)/len(female_df)}')
```

```python
full_df_no_id = full_df.drop('image_id',1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
cats = ['Male', 'Female', 'Pale', 'Unpale']

lengths = [len(male_df), len(female_df), len(pale_df), len(unpale_df)]
ax.bar(cats,lengths)
plt.show()
```
